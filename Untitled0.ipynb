```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(10)
data = pd.DataFrame({
    'value' : np.concatenate([np.random.normal(0, 1, 100), np.random.normal(10, 1, 10)])
})
Q1 = data['value'].quantile(0.25)
Q3 = data['value'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = data[(data['value'] < lower_bound) | (data['value'] > upper_bound)]
print(outliers)
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

data = {'age': [1, 2, 3, 4, 5],
        'height': [10, 20, 30, 40, 50]}
df = pd.DataFrame(data)

scaler = MinMaxScaler()

normalized_data = scaler.fit_transform(df)

normalized_df = pd.DataFrame(normalized_data, columns=df.columns)

normalized_df
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

data = {'age': [1, 2, 3, 4, 5],
        'height': [10, 20, 30, 40, 50]}
df = pd.DataFrame(data)

scaler = StandardScaler()

standerized_data = scaler.fit_transform(df)

standerized_df = pd.DataFrame(standerized_data, columns=df.columns)

print(standerized_df)

print("\nMean of standardized DataFrame:")
print(standerized_df.mean())

print("\nStandard Deviation of standardized DataFrame:")
print(standerized_df.std())
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

data = {'color': ['red', 'green', 'blue', 'red', 'green'],
        'size': ['small', 'medium', 'large', 'small', 'medium'],
        'price': [10, 20, 30, 10, 20]}
df = pd.DataFrame(data)

encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore') # sparse=False for dense output

columns_to_encode = ['color', 'size']


encoded_data = encoder.fit_transform(df[columns_to_encode])

encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(columns_to_encode))

df = pd.concat([df, encoded_df], axis=1)
df
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

data = {'customer_id': [1, 2, 3, 4, 5],
        'gender': ['Male', 'Female', 'Male', 'Female', 'Male'],
        'subscription': ['Basic', 'Premium', 'Basic', 'Premium', 'Free']}
df = pd.DataFrame(data)

label_encoder = LabelEncoder()

df['gender_encoded'] = label_encoder.fit_transform(df['gender'])

df['subscription_encoded'] = label_encoder.fit_transform(df['subscription'])

df
```
